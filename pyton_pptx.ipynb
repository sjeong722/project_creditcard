```python
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor
from pptx.enum.text import PP_ALIGN
from pptx.enum.shapes import MSO_SHAPE

# -----------------------------------------------------------------------------
# 1. 색상 상수 정의
# -----------------------------------------------------------------------------
ORANGE_MAIN   = RGBColor(255, 102, 0)   # #FF6600
ORANGE_LIGHT  = RGBColor(255, 153, 51)  # #FF9933
ORANGE_DARK   = RGBColor(204, 82, 0)    # #CC5200
WHITE         = RGBColor(255, 255, 255) # #FFFFFF
GREY_DARK     = RGBColor(51, 51, 51)    # #333333
GREY_LIGHT    = RGBColor(240, 240, 240) # #F0F0F0

# -----------------------------------------------------------------------------
# 2. 헬퍼 함수 정의 (수정됨)
# -----------------------------------------------------------------------------
def set_background(slide, color):
    """슬라이드 배경색 설정"""
    background = slide.background
    fill = background.fill
    fill.solid()
    fill.fore_color.rgb = color

def add_title(slide, text, font_size=40, color=ORANGE_MAIN, is_bold=True):
    """
    [수정사항]
    기본 title placeholder를 찾는 대신, 
    직접 텍스트 박스를 생성하여(add_textbox) 제목을 배치합니다.
    """
    # 상단 위치 고정
    left = Inches(0.5)
    top = Inches(0.5)
    width = Inches(9)
    height = Inches(1.2)
    
    # 텍스트 박스 생성
    txBox = slide.shapes.add_textbox(left, top, width, height)
    tf = txBox.text_frame
    
    p = tf.paragraphs[0]
    p.text = text
    p.font.size = Pt(font_size)
    p.font.color.rgb = color
    p.font.bold = is_bold
    p.alignment = PP_ALIGN.LEFT

def add_content_text(slide, text, color=GREY_DARK, font_size=18):
    """본문 텍스트 추가"""
    left = Inches(0.5)
    top = Inches(1.8)
    width = Inches(9)
    height = Inches(5)
    
    txBox = slide.shapes.add_textbox(left, top, width, height)
    tf = txBox.text_frame
    tf.word_wrap = True
    
    p = tf.add_paragraph()
    p.text = text
    p.font.size = Pt(font_size)
    p.font.color.rgb = color
    p.line_spacing = 1.5

# -----------------------------------------------------------------------------
# 3. 프레젠테이션 생성 및 슬라이드 구성
# -----------------------------------------------------------------------------
prs = Presentation()

# 슬라이드 마스터 크기 설정 (와이드 스크린 16:9)
prs.slide_width = Inches(13.333)
prs.slide_height = Inches(7.5)

# 데이터 정의
slides_data = [
    {"type": "cover", "title": "CARD & CATCH TABLE", "sub": "2026.01.26 | 웅진XSPOONS"},
    {"type": "contents", "title": "CONTENTS", "content": "1. 프로젝트 개요\n2. Step 1: 소셜 데이터\n3. Step 2: 커버리지 검증\n4. Step 3: 갭 도출\n5. STP 전략\n6. 마케팅 믹스 4P"},
    {"type": "body", "title": "프로젝트 개요", "content": "• What: 데이터 기반 신용카드 상품 기획\n• Why: 시장 니즈와 캐치테이블 유저 데이터 결합\n• How: 3단계 데이터 분석 프로세스 적용"},
    {"type": "body", "title": "Step 1: 소셜 데이터", "content": "• 키워드 TOP 10 분석\n• 맛집 탐방 트렌드 및 연관 키워드 도출\n• 2030 세대의 소비 패턴 파악"},
    {"type": "body", "title": "Step 2: 커버리지 검증", "content": "• 시중 30개 주요 신용카드 혜택 분석\n• 외식/다이닝 카테고리 혜택 비교\n• 기존 상품의 한계점 식별"},
    {"type": "body", "title": "Step 3: 갭 도출", "content": "• 소비자 니즈 vs 기존 혜택 간의 격차(Gap) 확인\n• '흑백 카드' 컨셉 도출: 프리미엄 vs 대중성\n• 차별화 포인트 설정"},
    {"type": "body", "title": "STP 전략", "content": "• Segmentation: 미식 고관여층 세분화\n• Targeting: 2539 직장인 미식가\n• Positioning: '예약부터 결제까지 끊김없는 경험'"},
    {"type": "body", "title": "마케팅 믹스 4P", "content": "• Product: 캐치테이블 제휴 특화 카드\n• Price: 연회비 및 리워드 구조 설계\n• Place: 앱 내 신청 프로세스 최적화\n• Promotion: 런칭 프로모션 및 팝업 이벤트"},
    {"type": "body", "title": "디지털 마케팅", "content": "• 콘텐츠 마케팅: 미식 가이드 발행\n• SEO 전략: '맛집 카드' 검색어 점유\n• 인플루언서 협업: 다이닝 경험 공유 캠페인"},
    {"type": "body", "title": "브랜드 전략", "content": "• 브랜드 아이덴티티: Modern & Premium Orange\n• Customer Journey Map 설계\n• 터치포인트별 브랜드 경험 일관성 유지"},
    {"type": "body", "title": "마케팅 실행", "content": "• Phase 1: 인지도 확보 (런칭 1개월)\n• Phase 2: 가입 유도 (런칭 2~3개월)\n• Phase 3: 사용 활성화\n• Phase 4: 충성도 강화"},
    {"type": "end", "title": "THANK YOU", "sub": "데이터 기반 신용카드 상품 기획 프로젝트"}
]

for i, data in enumerate(slides_data):
    # 빈 레이아웃(6번) 사용
    slide_layout = prs.slide_layouts[6] 
    slide = prs.slides.add_slide(slide_layout)
    slide_num = i + 1
    
    # 1. 표지
    if data['type'] == 'cover':
        set_background(slide, ORANGE_MAIN)
        
        # 화이트 박스
        box = slide.shapes.add_shape(MSO_SHAPE.RECTANGLE, Inches(1), Inches(2.5), Inches(11.3), Inches(2.5))
        box.fill.solid()
        box.fill.fore_color.rgb = WHITE
        box.line.fill.background()
        
        tf = box.text_frame
        p = tf.paragraphs[0]
        p.text = data['title']
        p.font.size = Pt(54)
        p.font.color.rgb = ORANGE_MAIN
        p.font.bold = True
        p.alignment = PP_ALIGN.CENTER
        
        # 서브 텍스트
        sub_box = slide.shapes.add_textbox(Inches(1), Inches(5.5), Inches(11.3), Inches(1))
        sp = sub_box.text_frame.paragraphs[0]
        sp.text = data['sub']
        sp.font.size = Pt(24)
        sp.font.color.rgb = WHITE
        sp.alignment = PP_ALIGN.CENTER
        sp.font.bold = True

    # 2. 목차
    elif data['type'] == 'contents':
        set_background(slide, WHITE)
        add_title(slide, data['title'], color=ORANGE_MAIN)
        add_content_text(slide, data['content'], color=GREY_DARK, font_size=24)

    # 3. 본문
    elif data['type'] == 'body':
        if slide_num % 2 != 0:
            bg_color = WHITE
        else:
            bg_color = GREY_LIGHT
        
        set_background(slide, bg_color)
        add_title(slide, data['title'], color=ORANGE_MAIN)
        
        # 구분선
        line = slide.shapes.add_shape(MSO_SHAPE.RECTANGLE, Inches(0.5), Inches(1.5), Inches(12), Inches(0.05))
        line.fill.solid()
        line.fill.fore_color.rgb = ORANGE_MAIN
        line.line.fill.background()
        
        add_content_text(slide, data['content'], color=GREY_DARK)

    # 4. 엔딩
    elif data['type'] == 'end':
        set_background(slide, ORANGE_MAIN)
        
        title_box = slide.shapes.add_textbox(Inches(1), Inches(3), Inches(11.3), Inches(2))
        tp = title_box.text_frame.paragraphs[0]
        tp.text = data['title']
        tp.font.size = Pt(60)
        tp.font.color.rgb = WHITE
        tp.font.bold = True
        tp.alignment = PP_ALIGN.CENTER
        
        sub_box = slide.shapes.add_textbox(Inches(1), Inches(5), Inches(11.3), Inches(1))
        sp = sub_box.text_frame.paragraphs[0]
        sp.text = data['sub']
        sp.font.size = Pt(24)
        sp.font.color.rgb = WHITE
        sp.alignment = PP_ALIGN.CENTER

# -----------------------------------------------------------------------------
# 4. 저장
# -----------------------------------------------------------------------------
file_path = "CatchTable_Brand_Pitch_Fixed.pptx"
prs.save(file_path)
print(f"파일이 정상적으로 생성되었습니다: {file_path}")
```
