```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib import rcParams, font_manager as fm
import numpy as np
import platform

# ==================== 한글 폰트 설정 ====================
system = platform.system()
if system == 'Windows':
    font_name = 'Malgun Gothic'
elif system == 'Darwin':
    font_name = 'AppleGothic'
else:
    font_name = 'DejaVu Sans'

rcParams['font.family'] = font_name
plt.rcParams['axes.unicode_minus'] = False

# ==================== 전문가 수준 스타일 설정 ====================
plt.style.use('seaborn-v0_8-darkgrid')
rcParams['figure.facecolor'] = '#FFFFFF'
rcParams['axes.facecolor'] = '#F8F9FA'
rcParams['grid.color'] = '#E0E0E0'
rcParams['grid.linewidth'] = 0.5

# ==================== 데이터 설정 ====================
categories = ['배달앱', '카페', '온라인쇼핑', '편의점', 'OTT', '기타']
consumption_ratio = [25, 20, 15, 15, 10, 15]  # 20대 실제 지출 비중
z_life_discount = [12, 10, 8, 10, 20, 0]     # Z-LIFE 카드 할인율

# 월 40만원 기준 절감액
spending_40 = [100000, 80000, 60000, 60000, 40000, 60000]  # 카테고리별 월 소비
savings_amount = [12000, 8000, 4800, 6000, 8000, 0]  # 절감액

colors = ['#FF6B6B', '#FFA500', '#4ECDC4', '#45B7D1', '#96CEB4', '#BDBDBD']
colors_light = ['#FFB8B8', '#FFD4A6', '#A9E5DF', '#A6D5F7', '#D5EFDF', '#E8E8E8']

# ==================== 그래프 생성 (전문 수준) ====================
fig = plt.figure(figsize=(18, 10))
gs = fig.add_gridspec(3, 2, hspace=0.35, wspace=0.3, 
                       left=0.08, right=0.95, top=0.93, bottom=0.08)

# 메인 제목
fig.suptitle('20대 소비 현황 분석 & Z-LIFE 카드 절감효과 비교', 
             fontsize=22, fontweight='bold', y=0.98, color='#1a1a1a')

# ==================== 좌측: 20대 소비 현황 ====================
ax1 = fig.add_subplot(gs[0:2, 0])

x = np.arange(len(categories))
width = 0.35

bars1 = ax1.bar(x - width/2, consumption_ratio, width, 
                label='20대 실제 지출 비중', 
                color=colors, alpha=0.85, edgecolor='white', linewidth=2)

bars2 = ax1.bar(x + width/2, z_life_discount, width,
                label='Z-LIFE 카드 할인율', 
                color=colors_light, alpha=0.85, edgecolor='white', linewidth=2)

# 값 라벨 (상단)
for bar in bars1:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.5,
            f'{int(height)}%', ha='center', va='bottom', 
            fontsize=11, fontweight='bold', color='#333333')

for bar in bars2:
    height = bar.get_height()
    if height > 0:
        ax1.text(bar.get_x() + bar.get_width()/2., height + 0.5,
                f'{int(height)}%', ha='center', va='bottom', 
                fontsize=11, fontweight='bold', color='#333333')

ax1.set_ylabel('비중 / 할인율 (%)', fontsize=13, fontweight='bold', color='#333333')
ax1.set_title('그래프 1: 20대 월별 소비 현황\n(월평균 40만원 기준)', 
             fontsize=14, fontweight='bold', pad=15, color='#1a1a1a')
ax1.set_xticks(x)
ax1.set_xticklabels(categories, fontsize=11, fontweight='bold')
ax1.legend(fontsize=11, loc='upper right', framealpha=0.95, 
          edgecolor='#CCCCCC', fancybox=True, shadow=True)
ax1.set_ylim(0, 28)
ax1.grid(axis='y', alpha=0.3, linestyle='--', linewidth=0.8)
ax1.set_axisbelow(True)

# 데이터 출처 (좌측 하단)
ax1.text(0.02, 0.02, '자료: 대학내일20대연구소 공식조사 (2025년 9월)',
        transform=ax1.transAxes, fontsize=9, style='italic', 
        color='#666666', bbox=dict(boxstyle='round', facecolor='#F5F5F5', alpha=0.7, pad=0.5))

# ==================== 우측: 절감 효과 (이중 축) ====================
ax2 = fig.add_subplot(gs[0:2, 1])
ax2_twin = ax2.twinx()

# 소비액 (막대)
bars_consumption = ax2.bar(x - 0.2, spending_40, 0.4, 
                          label='월 소비액', color='#E3F2FD', 
                          alpha=0.7, edgecolor='#1976D2', linewidth=1.5)

# 절감액 (막대)
bars_savings = ax2.bar(x + 0.2, savings_amount, 0.4,
                       label='월 절감액', color=colors, 
                       alpha=0.85, edgecolor='white', linewidth=2)

# 할인율 (꺾은선)
line = ax2_twin.plot(x, z_life_discount, marker='o', linewidth=3, 
                    markersize=10, color='#2ECC71', label='할인율',
                    markerfacecolor='#27AE60', markeredgecolor='white', 
                    markeredgewidth=2)

# 값 라벨
for bar in bars_savings:
    height = bar.get_height()
    if height > 0:
        ax2.text(bar.get_x() + bar.get_width()/2., height + 200,
                f'{int(height):,}원', ha='center', va='bottom', 
                fontsize=10, fontweight='bold', color='#d32f2f')

for i, rate in enumerate(z_life_discount):
    if rate > 0:
        ax2_twin.text(i, rate + 1.5, f'{int(rate)}%', ha='center', 
                     fontsize=10, fontweight='bold', color='#2ECC71')

ax2.set_ylabel('소비액 / 절감액 (원)', fontsize=13, fontweight='bold', color='#333333')
ax2_twin.set_ylabel('할인율 (%)', fontsize=13, fontweight='bold', color='#2ECC71')
ax2.set_title('그래프 2: Z-LIFE 카드 이용 시 절감효과\n(월소비 40만원 시나리오)', 
             fontsize=14, fontweight='bold', pad=15, color='#1a1a1a')
ax2.set_xticks(x)
ax2.set_xticklabels(categories, fontsize=11, fontweight='bold')
ax2.set_ylim(0, 120000)
ax2_twin.set_ylim(0, 25)

# 범례
lines1, labels1 = ax2.get_legend_handles_labels()
lines2, labels2 = ax2_twin.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc='upper left', 
          fontsize=11, framealpha=0.95, edgecolor='#CCCCCC', 
          fancybox=True, shadow=True)

ax2.grid(axis='y', alpha=0.3, linestyle='--', linewidth=0.8)
ax2.set_axisbelow(True)
ax2.yaxis.label.set_color('#333333')
ax2_twin.yaxis.label.set_color('#2ECC71')
ax2.tick_params(axis='y', labelcolor='#333333')
ax2_twin.tick_params(axis='y', labelcolor='#2ECC71')

# ==================== 하단: 핵심 지표 요약 ====================
ax3 = fig.add_subplot(gs[2, :])
ax3.axis('off')

# 요약 텍스트 박스들
summary_data = [
    ('월 총 절감액', '28,000원', '#FF6B6B'),
    ('연간 절감액', '336,000원', '#FFA500'),
    ('연회비', '무료', '#4ECDC4'),
    ('최소 실적', '30만원', '#45B7D1'),
    ('카드 주요 혜택', '배달 12% · 카페 10% · OTT 20%', '#96CEB4'),
]

box_width = 0.18
start_x = 0.05

for i, (label, value, color) in enumerate(summary_data):
    x_pos = start_x + (i * 0.19)
    
    # 박스 배경
    box = mpatches.FancyBboxPatch((x_pos, 0.25), box_width, 0.65,
                                 boxstyle='round,pad=0.01',
                                 transform=ax3.transAxes,
                                 facecolor=color, alpha=0.15,
                                 edgecolor=color, linewidth=2.5)
    ax3.add_patch(box)
    
    # 라벨
    ax3.text(x_pos + box_width/2, 0.75, label, 
            transform=ax3.transAxes, ha='center', va='top',
            fontsize=11, fontweight='bold', color='#333333')
    
    # 값
    ax3.text(x_pos + box_width/2, 0.45, value,
            transform=ax3.transAxes, ha='center', va='center',
            fontsize=12, fontweight='bold', color=color)

# 하단 주석
fig.text(0.5, 0.02, 
        '© 2025 Z-LIFE 신용카드 상품 분석 | 데이터: 대학내일20대연구소(2025년 9월) | 신용카드 시장 통계(2025년)',
        ha='center', fontsize=9, color='#999999', style='italic')

plt.savefig('20s_credit_card_analysis_professional.png', 
           dpi=300, bbox_inches='tight', facecolor='white', edgecolor='none')
print("✅ 전문 수준의 그래프 생성 완료: 20s_credit_card_analysis_professional.png")
plt.show()
```
