# 파생변수 만들기

```python
# 총이용금액_R3M
df2_통합['총이용금액_R3M'] = (
    df2_통합[['이용금액_일시불_R3M','이용금액_할부_유이자_R3M','이용금액_CA_R3M','이용금액_카드론_R3M']].sum(axis=1)
)

# 총이용금액_R3M 음수값 제거
# 승인건의 취소,환불,정정,선결제,초과납부일 경우 음수로 집계될 수 있음
df2_통합 = df2_통합[df2_통합['총이용금액_R3M']>0]
```

```python
# 수익
df2_통합['수익'] = (
    (df2_통합['이용금액_일시불_R3M']*0)
    + (df2_통합['이용금액_할부_유이자_R3M']*0.1)
    + (df2_통합['이용금액_CA_R3M']*0.15)
    + (df2_통합['이용금액_카드론_R3M']*0.1)
)
```

```python
# 수익률
df2_통합['수익률'] = (
    (df2_통합['수익']) - (df2_통합['혜택수혜금액_R3M'])
    / df2_통합['총이용금액_R3M']
)
```

```python
# 이상치 탐지

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# 한글 폰트 설정 (Colab 환경 등에서 필요)
plt.rcParams['font.family'] = 'NanumGothic' # 또는 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

def visualize_outliers(df, column, weight=1.5):
    """
    IQR 방식으로 이상치를 탐지하고 시각화합니다.
    """
    # 1. IQR 및 경계값 계산
    quantile_25 = np.percentile(df[column].values, 25)
    quantile_75 = np.percentile(df[column].values, 75)
    iqr = quantile_75 - quantile_25
    iqr_weight = iqr * weight

    lowest_val = quantile_25 - iqr_weight
    highest_val = quantile_75 + iqr_weight

    # 2. 이상치 인덱스 추출
    outlier_index = df[(df[column] < lowest_val) | (df[column] > highest_val)].index

    # 3. 시각화 (Boxplot & Histogram)
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # 왼쪽: Boxplot (중앙값, 사분위수, 이상치 점을 확인)
    sns.boxplot(x=df[column], ax=axes[0], color='skyblue')
    axes[0].axvline(lowest_val, color='red', linestyle='--', label=f'Lower Bound: {lowest_val:,.0f}')
    axes[0].axvline(highest_val, color='red', linestyle='--', label=f'Upper Bound: {highest_val:,.0f}')
    axes[0].set_title(f'[{column}] Boxplot', fontsize=14)
    axes[0].legend()

    # 오른쪽: Histogram (데이터의 전반적인 분포와 꼬리 확인)
    sns.histplot(df[column], kde=True, ax=axes[1], color='orange')
    axes[1].axvline(lowest_val, color='red', linestyle='--')
    axes[1].axvline(highest_val, color='red', linestyle='--')
    axes[1].set_title(f'[{column}] Distribution', fontsize=14)

    plt.suptitle(f'Outlier Detection for {column}', fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

    # 4. 통계 정보 출력
    print(f"--- {column} 이상치 분석 결과 ---")
    print(f"전체 데이터 개수: {len(df):,}")
    print(f"이상치 개수: {len(outlier_index):,}")
    print(f"이상치 비율: {len(outlier_index)/len(df)*100:.2f}%")
    print(f"하한 경계값(Lower Bound): {lowest_val:,.2f}")
    print(f"상한 경계값(Upper Bound): {highest_val:,.2f}")

    return outlier_index


# --- 적용 예시 ---
# 1. 특정 컬럼('이용금액_일시불_R3M') 시각화
outlier_idx = visualize_outliers(df2_통합, '이용금액_일시불_R3M')

# 2. 여러 수치형 컬럼에 대해 반복해서 확인하고 싶을 때
# check_cols = ['이용금액_일시불_R3M', '이용금액_할부_유이자_R3M', '총이용금액_R3M']
# for col in check_cols:
#     visualize_outliers(df2_통합, col)
```

```python
# 이상치 처리 방법 B: 경계값 대체 (Capping)

def cap_outliers(df, column, weight=1.5):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - (iqr * weight)
    upper_bound = q3 + (iqr * weight)

    # 경계값을 벗어나는 데이터를 경계값으로 대체
    df[column] = np.where(df[column] > upper_bound, upper_bound,
                 np.where(df[column] < lower_bound, lower_bound, df[column]))
    return df

# 적용
df1_매출 = cap_outliers(df1_매출, '이용금액_일시불_R3M')
```
